 this assignment shows following operations on ‘cars_data’ dataset:
    
1. Read the dataset to the python environment.

# 1.Read the dataset to python enviornment

```python
import pandas as pd
df=pd.read_csv('cars_data.csv')
df
```

# 2.Check for the null values present in the dataset

```python
df.isna().sum()
```

```python
df.info()
```

# 3.Plot a bar graph of male vs female buyers participated in the sales.

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6))
df['Buyer Gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender", fontsize='15')
plt.ylabel("Total count", fontsize='15')
```

# Insights

#### Most highlighting fact in this analysis is no.of Female car buyers is more than no.of Male car buyers

# 4. Find the top 5 cars based on their sales price.

```python
top5 = df.sort_values(by='Sale Price',ascending=False)
top5.reset_index(drop=True).head(5)
```

```python
top5['Make'].reset_index(drop=True).head(5)
```

# Insights

these are the 5 cars having highest sales price in descending order

# 5.Find the least 5 cars based on their Resell price.

```python
least5= df.sort_values(by='Resell Price',ascending=True)
least5.reset_index(drop=True).head(5)
```

```python
least5['Make'].reset_index(drop=True).head(5)
```

# insight

These are the 5 cars having lowest resell price in ascending order
